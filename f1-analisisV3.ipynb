# Tarea 1
Crear el scrip que genere 4 archivos csv que contengan la información de las siguientes tablas:

### Drivers

```python
drivers ={'year':[],'driverId':[],'number':[],'code':[],'forname':[],'sumame':[],'dob':[],'nationality':[],'url':[]}
for year in list(range(2010,2023)):
    url='http://ergast.com/api/f1/{}/drivers.json'
    response=requests.get(url.format(year))
    content=json.loads(response.content)
    for item in content['MRData']['DriverTable']['Drivers']:
        drivers['year'].append(year)
        drivers['driverId'].append(item['driverId'])
        try:
            drivers['number'].append(int(item['permanentNumber']))
        except:
            drivers['number'].append(None)
        drivers['code'].append(item['code'])
        drivers['forname'].append(item['familyName'])
        drivers['sumame'].append(item['givenName'])
        drivers['dob'].append(item['dateOfBirth'])
        drivers['nationality'].append(item['nationality'])
        drivers['url'].append(item['url'])
drivers['driverRef']=np.arange(1,pd.DataFrame(drivers).shape[0]+1,1)
pd.DataFrame(drivers)
```

```python
pd.DataFrame(drivers).to_csv('drivers.csv',index=False)
```

```python
qualifying ={'year':[],'qualifyingId':[],'raceId':[],'driverId':[],'constructorId':[],'number':[],'position':[],'Q1':[],'Q2':[],'Q3':[]}
for year in range(2010,2023):
    url= 'http://ergast.com/api/f1/{}/qualifying.json'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    
    for item in content['MRData']['RaceTable']['Races'][0]['QualifyingResults']:
        qualifying['year'].append(year)
        #qualifying['qualifyingId'].append(item['qualifyingId'])
        qualifying['raceId'].append(content['MRData']['RaceTable']['Races'][0]['raceName'])
        qualifying['driverId'].append(item['Driver']['driverId'])
        qualifying['constructorId'].append(item['Constructor']['constructorId'])
        qualifying['number'].append(int(item['number']))
        qualifying['position'].append(int(item['position']))
        qualifying['Q1'].append(item['Q1'])
        try:
            qualifying['Q2'].append(item['Q2'])
        except:
            qualifying['Q2'].append(None)
        try:
            qualifying['Q3'].append(item['Q3'])
        except:
            qualifying['Q3'].append(None)
    for item in content['MRData']['RaceTable']['Races'][1]['QualifyingResults']:
        qualifying['year'].append(year)
        #qualifying['qualifyingId'].append(item['qualifyingId'])
        qualifying['raceId'].append(content['MRData']['RaceTable']['Races'][1]['raceName'])
        qualifying['driverId'].append(item['Driver']['driverId'])
        qualifying['constructorId'].append(item['Constructor']['constructorId'])
        qualifying['number'].append(int(item['number']))
        qualifying['position'].append(int(item['position']))
        qualifying['Q1'].append(item['Q1'])
        try:
            qualifying['Q2'].append(item['Q2'])
        except:
            qualifying['Q2'].append(None)
        try:
            qualifying['Q3'].append(item['Q3'])
        except:
            qualifying['Q3'].append(None)
qualifying['qualifyingId']=np.arange(1,391,1)
pd.DataFrame(qualifying)
```

### sprintResults

```python
sprintResult ={'year':[],'sprintResultId':[],'raceId':[],'driverId':[],'constructorId':[],'number':[],'grid':[],
               'position':[],'positionText':[],'points':[],'laps':[],'time':[],'milliseconds':[],
               'fastestLap':[],'fastestLapTime':[],'statusId':[]}
for year in range(2010,2023):
    url= 'http://ergast.com/api/f1/{}/sprint.json?limit=1000'
    response = requests.get(url.format(year))
    content=json.loads(response.content)
    inicio=content['MRData']['RaceTable']
    for item in range(0,len(inicio['Races'])):
        for pivote in inicio['Races'][item]['SprintResults']:
            sprintResult['year'].append(year)
            sprintResult['raceId'].append(inicio['Races'][item]['raceName'])
            sprintResult['driverId'].append(pivote['Driver']['driverId'])
            sprintResult['constructorId'].append(pivote['Constructor']['constructorId'])
            sprintResult['number'].append(int(pivote['number']))
            sprintResult['grid'].append(pivote['grid'])
            sprintResult['position'].append(int(pivote['position']))
            sprintResult['positionText'].append((pivote['positionText']))
            sprintResult['points'].append(int(pivote['points']))
            sprintResult['laps'].append(int(pivote['laps']))
            try:
                sprintResult['time'].append(pivote['Time']['time'])
            except:
                sprintResult["time"].append(None)
            try:
                sprintResult['milliseconds'].append(pivote['Time']['millis'])
            except:
                sprintResult["milliseconds"].append(None)
            try:
                sprintResult['fastestLap'].append(int(pivote['FastestLap']['lap']))
            except:
                sprintResult["fastestLap"].append(None)
            try:
                sprintResult['fastestLapTime'].append(pivote['Time']['time'])
            except:
                sprintResult["fastestLapTime"].append(None)
            sprintResult['statusId'].append(pivote['status'])
sprintResult['sprintResultId']=np.arange(1,121,1)
pd.DataFrame(sprintResult)
```

### pitStops

```python
pitStops = {'year':[],"raceId":[],'driverId':[], 'stop':[],
           'lap':[], 'time':[], 'duration':[]}#, 'milliseconds':[]}

for year in range(2010,2023):
    url='http://ergast.com/api/f1' 
    url1=url+'/{}.json?limit=1000'
    url3=url+'/{}'.format(year)
    response = requests.get(url1.format(year))
    content=json.loads(response.content)
    tampath=len(content['MRData']['RaceTable']['Races'])
    links=[]
    for i in range(1,tampath+1):
        url2=url3+'/{}/pitstops.json?limit=1000'
        response = requests.get(url2.format(i))
        content=json.loads(response.content)
        links.append(url2.format(i))
    i=0 
    for j in links:
        try:
            response = requests.get(j)
            content=json.loads(response.content)
        except:
            print(f"Hay un error en {j}")
        try: 
            content['MRData']['RaceTable']['Races'][0]['PitStops']
            for item in content['MRData']['RaceTable']['Races'][0]['PitStops']:
                pitStops["raceId"].append(content['MRData']['RaceTable']['Races'][0]['round'])
                pitStops["year"].append(content['MRData']['RaceTable']['Races'][0]['season'])
                pitStops["driverId"].append(item['driverId'])
                pitStops["stop"].append(item['stop'])
                pitStops["lap"].append(item['lap'])
                pitStops["time"].append(item['time'])
                pitStops["duration"].append(item['duration'])
            
        except:
            print(f"Hay un error en {j}")
            
        

pitStopss=pd.DataFrame(pitStops)         
pitStopss.head()
```

### LapTimes

```python
lapTimes={"year":[],"round":[],"raceId":[],"driverId":[],"lap":[],"position":[],"time":[]}

for year in range(2010,2023):
    url='http://ergast.com/api/f1' 
    url1=url+'/{}.json?limit=1000'
    url3=url+'/{}'.format(year)
    response = requests.get(url1.format(year))
    content=json.loads(response.content)
    tampath=len(content['MRData']['RaceTable']['Races'])
    links=[]
    for i in range(1,tampath+1):
        url2=url3+'/{}/laps.json?limit=10000'
        response = requests.get(url2.format(i))
        content=json.loads(response.content)
        links.append(url2.format(i))
    for j in links:
        print(j)
        response = requests.get(j)
        content=json.loads(response.content)
        for i in content['MRData']['RaceTable']["Races"][0]["Laps"]:
            for item in i['Timings']:
                try:
                    lapTimes["year"].append(content['MRData']['RaceTable']['Races'][0]['season'])                    
                except:
                    lapTimes["year"].append("")
                try:
                    lapTimes["round"].append(content['MRData']['RaceTable']['Races'][0]['round'])                    
                except:
                    lapTimes["round"].append("")
                try:
                    lapTimes["raceId"].append(content['MRData']['RaceTable']["Races"][0]['raceName'])
                except:
                    lapTimes["raceId"].append("")
                try:
                    lapTimes["driverId"].append(item['driverId'])
                except:
                    lapTimes["driverId"].append("")
                try:
                    lapTimes["lap"].append(i['number'])
                except:
                    lapTimes["lap"].append("")
                try:
                    lapTimes["position"].append(item['position'])
                except:
                    lapTimes["position"].append("")
                try:
                    lapTimes["time"].append(item['time'])
                except:
                    lapTimes["time"].append("")
                
lapTimess=pd.DataFrame.from_dict(lapTimes,orient='index')
lapTimess=lapTimess.transpose()
```
